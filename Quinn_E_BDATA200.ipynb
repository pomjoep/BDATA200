## Exercise: Functions

```
def celsius_to_fahrenheit(c : float) -> float:
    return (c * 9 / 5) + 32;

for item in [0, 100, 37]:
    print(f"{item} celsius is {celsius_to_fahrenheit(item)} in fahrenheit");
```

## Exercise: NumPy Arrays

```
import numpy as np;

arr = np.array([15, 22, 8, 31, 17]);
print(arr.mean());
print(arr.max());
print(arr.std());

arr * 3;
print(f"\n{arr}");
```

## Exercise: pandas Basics

```
import pandas as pd;

df = pd.DataFrame({
    "name": ["Seattle", "Los Angeles", "Chicago"],
    "population": [816000, 3900000, 2700000] 
})

print(df.shape);
print(df.dtypes);
print(df.describe());

maxIndex = df["population"].idxmax();
print(f"\nThe most populous city in this list is {df.loc[maxIndex, 'name']}");
```
