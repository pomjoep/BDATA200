## Exercise: matplotlib

```python
import matplotlib.pyplot as plt;
import pandas as pd;

auto = pd.read_csv("https://raw.githubusercontent.com/PedroBSB/BDATA200/refs/heads/main/Data/Auto.csv", na_values="?");


fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 10));
ax[0, 0].scatter(auto["horsepower"], auto["acceleration"]);
ax[0, 0].set_xlabel("horsepower");
ax[0, 0].set_ylabel("acceleration 0-60 in seconds");
ax[0, 0].set_title("Horsepower vs Acceleration");
ax[0, 0].invert_yaxis();

ax[0, 1].hist(auto["weight"], bins = 15);
ax[0, 1].set_xlabel("weight");
ax[0, 1].set_ylabel("number of cars");
ax[0, 1].set_title("Weight Distribution of Cars");

originDictionary = {
    1: "America",
    2: "Europe",
    3: "Asia"
};
auto["origin"] = auto["origin"].replace(originDictionary);
auto.groupby("origin")["horsepower"].mean().plot(kind="bar", ax = ax[1, 0]);
ax[1, 0].set_ylabel("Average Horsepower");
ax[1, 0].set_xlabel("Car Region");
ax[1, 0].set_title("Average Horsepower by Region");

auto.groupby("year")["acceleration"].mean().plot(kind="line", ax = ax[1, 1]);
ax[1, 1].set_ylabel("Average Acceleration 0-60 in seconds");
ax[1, 1].set_xlabel("Year");
ax[1, 1].set_title("Average Acceleration by Year");
ax[1, 1].invert_yaxis();

plt.show();
```

## Exercise: Plotnine

```python
from plotnine import *;
import pandas as pd;    

auto = pd.read_csv("https://raw.githubusercontent.com/PedroBSB/BDATA200/refs/heads/main/Data/Auto.csv", na_values="?");

plot = (ggplot(auto, aes("displacement", "mpg"))
         + geom_point(alpha = 0.5, color="#4B2E83")
         + labs(title = "MPG vs Displacement", x = "Displacement", y = "MPG")
         + theme_classic()
         );

plot2 = (ggplot(auto, aes("acceleration", fill="factor(origin)"))
         + geom_histogram(bins = 20, color = "white", alpha = 0.5)
         + labs(title="Acceleration Distribution by Origin", x="Acceleration", y="Count", fill="Origin")
         + theme_minimal()
         );

plot3 = (ggplot(auto, aes(x = "factor(origin)", y = "weight", fill = "factor(origin)"))
         + geom_errorbar(stat="boxplot", width=0.2)
         + geom_boxplot(alpha = 0.5)
         + labs(title="Weight by Origin", x="Origin", y="Weight", fill="Origin")
         + theme_dark()
         );

display(plot, plot2, plot3);
```
