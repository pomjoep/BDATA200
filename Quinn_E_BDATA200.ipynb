## Exercise: Functions

```
def celsius_to_fahrenheit(c : float) -> float:
    return (c * 9 / 5) + 32;

for item in [0, 100, 37]:
    print(f"{item} celsius is {celsius_to_fahrenheit(item)} in fahrenheit");
```

## Exercise: pandas Basics

```
import pandas as pd;

df = pd.DataFrame({
    "name": ["Seattle", "Los Angeles", "Chicago"],
    "population": [816000, 3900000, 2700000]
})

print(df.shape);
print(df.dtypes);
print(df.describe());

maxIndex = df["population"].idxmax();
print(f"\nThe most populous city in this list is {df.loc[maxIndex, 'name']}");
```

## Exercise: Reading & Exploring Date

```
import pandas as pd;

url = "https://raw.githubusercontent.com/PedroBSB/BDATA200/refs/heads/main/Data/Auto.csv";
auto = pd.read_csv(url, na_values="?");

print(f"Rows in set: {auto.shape[0]} \nColumns in set: {auto.shape[1]}");

print(f"Average MPG: {auto['mpg'].mean()}");
print(f"Max HP: {auto['horsepower'].max()}, from the: {auto.loc[auto['horsepower'].idxmax(), 'name']}");
```

## Exercise: matplotlib

```
import matplotlib.pyplot as plt;
import pandas as pd;

auto = pd.read_csv("https://raw.githubusercontent.com/PedroBSB/BDATA200/refs/heads/main/Data/Auto.csv", na_values="?");


fig, ax = plt.subplots();
ax.scatter(auto["horsepower"], auto["acceleration"]);
ax.set_xlabel("horsepower");
ax.set_ylabel("acceleration 0-60 in seconds");
ax.set_title("Horsepower vs Acceleration");
ax.invert_yaxis();
plt.show();

ax.hist(auto["weight"], bins = 15);
ax.set_xlabel("weight");
ax.set_ylabel("number of cars");
ax.set_title("Weight Distribution of Cars");

```
